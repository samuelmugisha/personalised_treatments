```python
import math
from matplotlib import pyplot as plt

Vd = 3.2
def generate_first_order(Dosage):
    ke = math.log(2)/(20*60)
    ke1 = math.log(2)/(2.3*60)
    ke2 = math.log(2)/(20)
    Co = Dosage/Vd
    i = 0
    Ct = Co * math.pow(2.718,-ke*i)
    x = [i for i in range(0,3600)]
    y = [Co * math.pow(2.718,-ke*i) for i in x]
    w = [Co * math.pow(2.718,-ke1*i) for i in x]
    z = [Co * math.pow(2.718,-ke2*i) for i in x]
    img = plt.plot(x,y, label="doxorubicin")
    img2 = plt.plot(x,w, label="vincristine")
    img3 = plt.plot(x,z, label="cisplatin")
    plt.legend()
    plt.title("Plasma concentration with time")
    plt.xlabel("Time")
    plt.ylabel("C")
    plt.show()
    return img, img2,img3


print (generate_first_order(100))
```
