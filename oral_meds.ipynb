```python
import math
import numpy as np
from matplotlib import pyplot as plt


def generate_one_compartment_oral(halfLife,Dosage):
    Vd = 3.2
    Ke = math.log(2)/(halfLife*60)
    print(Ke)
    Co = Dosage/Vd
    Ka = 0.0001
    Ka2 = 0.00001
    Ka3 = 0.000001
    
    i = 0
    x = [i for i in range(0,13600)]
    y = [ Co*Ka*(math.pow(2.718,-Ke*i)-math.pow(2.718,-Ka*i))/(Ka-Ke) for i in x]
    z = [ Co*Ka2*(math.pow(2.718,-Ke*i)-math.pow(2.718,-Ka2*i))/(Ka2-Ke) for i in x]
    w = [ Co*Ka3*(math.pow(2.718,-Ke*i)-math.pow(2.718,-Ka3*i))/(Ka3-Ke) for i in x]
    img2 = plt.plot(x,y, label = "Amoxicilin 250mg cap")
    plt.plot(x,z, label="Penicillin V 250mg tab")
    plt.plot(x,w, label="Griseofulvin 500mg  tab")
    plt.legend()
    plt.title("Plasma concentration with time")
    plt.xlabel("Time")
    plt.ylabel("C")
    plt.show()
    return img2

print(generate_one_compartment_oral(10,30))
```
